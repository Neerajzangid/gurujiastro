```python
top_earners = df.groupby('guruName')['astrologersEarnings'].sum().nlargest(5).reset_index()
plt.figure(figsize=(12, 8))

plt.figure(figsize=(12, 8))
bar_plot = sb.barplot(x='astrologersEarnings', y='guruName', data=top_earners, palette='coolwarm')

# Add annotations on the bars
for index, value in enumerate(top_earners['astrologersEarnings']):
    bar_plot.text(value, index, f'{value:,.2f}', va='center', ha='left', size=12, color='black', weight='bold')

plt.title('Top 5 Guruji by Earnings', fontsize=20, fontweight='bold')
plt.xlabel('Total Earnings', fontsize=15)
plt.ylabel('Guruji', fontsize=15)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.grid(True, linestyle='--', alpha=0.6)

# Remove top and right spines for a cleaner look
sb.despine()

plt.show()



```

```python
#Please create a line chart showing the trend in call charges over time.
bins = list(range(0, 75, 5))  # Create bins of 5-second intervals
labels = [f'{i}-{i+5}' for i in range(0, 70, 5)]  # Ensure one fewer label than the number of bin edges

df['time_bins'] = pd.cut(df['timeDuration'], bins=bins, labels=labels, right=False)

time_bins_charges = df.groupby('time_bins')['amount'].sum().reset_index()

plt.figure(figsize=(12, 6))
plt.plot(time_bins_charges['time_bins'], time_bins_charges['amount'], marker='o', linestyle='-')
plt.title('Trend in Call Charges Over Time Intervals')
plt.xlabel('Time Intervals (seconds)')
plt.ylabel('Total Call Charges')
plt.grid(True)
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sb.scatterplot(x='timeDuration', y='amount', data=df)
plt.title('Relationship between timeDuration and Amount')
plt.xlabel('timeDuration')
plt.xlim(0,100)
plt.ylabel('Amount')
plt.ylim(0,2000)
plt.grid(True)
plt.show()
```
